```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor
from math import sqrt
```

```python
df=pd.read_excel('combine-ecw.xlsx')
df1 = df.copy()
df1 = df.drop(['ILs', 'density (g/cm3)', 'viscosity (P)', 'Anion_LUMO (eV)', 'Anion_dipole_mo(D)', 'MF', 'avg_valence_p_elec.'], axis=1)
x = df1.drop(["ECW (V)"], axis=1)
y = df1["ECW (V)"]
```

```python
xgb = XGBRegressor(objective='reg:squarederror', random_state=42)
```

## using best parameters

```python
xgb = XGBRegressor(base_score=1, learning_rate=0.15, max_depth=20, min_child_weight=3, n_estimators=100, objective='reg:squarederror', random_state=15, verbose=20)
```

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import mean_squared_error
```

```python
k_folds = KFold(n_splits=10)
scores = cross_val_score(xgb,x,y,cv=k_folds, scoring='neg_mean_absolute_error')
print(scores)
print(np.mean(scores))
```

## from milan

```python
from sklearn.utils import shuffle
k_folds = KFold(n_splits=5)
X_shuffle, y_shuffle = shuffle(x, y, random_state=42)
from sklearn.model_selection import cross_val_score
scores = cross_val_score(xgb, X_shuffle, y_shuffle,
                         scoring="neg_mean_squared_error",
                         cv=k_folds, n_jobs=1)
#scores = cross_val_score(xgb,x,y,cv=k_folds, scoring='neg_mean_absolute_error')
rmse = np.sqrt(-scores)
print("RMSE values: ", np.round(rmse, 3))
print("RMSE average: ", np.mean(rmse))
```
